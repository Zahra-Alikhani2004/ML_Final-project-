```python
# ============================
# Iran Futsal Players Dataset Analysis - Complete Version with Extra Charts
# ============================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set(style="whitegrid")

# 1Ô∏è‚É£ Load Dataset
url = "https://raw.githubusercontent.com/Zahra-Alikhani2004/ML_Final-project-/main/Futsal-IRAN-Dataset/futsal_iran_players.csv"
df = pd.read_csv(url)

# 2Ô∏è‚É£ Rename Columns to English
df.rename(columns={
    'ŸÜÿßŸÖ':'Name',
    'ŸÇÿØ (ŸÖÿ™ÿ±)':'Height_m',
    'Ÿàÿ≤ŸÜ (⁄©€åŸÑŸà⁄Øÿ±ŸÖ)':'Weight_kg',
    'Ÿæÿ≥ÿ™':'Position',
    'ÿ≥ŸÜ ÿ¥ÿ±Ÿàÿπ ÿ≠ÿ±ŸÅŸá‚Äåÿß€å':'Pro_Start_Age',
    'ÿ≥ÿ±ÿπÿ™':'Speed',
    '⁄Üÿßÿ®⁄©€å':'Agility',
    'ÿßÿ≥ÿ™ŸÇÿßŸÖÿ™':'Endurance',
    'ŸÖŸáÿßÿ±ÿ™ Ÿæÿßÿ≥':'Passing',
    'ŸÖŸáÿßÿ±ÿ™ ÿØÿ±€åÿ®ŸÑ':'Dribbling',
    'ŸÖŸáÿßÿ±ÿ™ ÿ¥Ÿàÿ™':'Shooting'
}, inplace=True)

# 3Ô∏è‚É£ Translate Positions to English
position_translation = {
    'ŸÖŸáÿßÿ¨ŸÖ': 'Forward',
    'Pivot': 'Pivot',
    'ŸÅŸÑŸÜ⁄© ÿ±ÿßÿ≥ÿ™': 'Right Flank',
    'ŸÅŸÑŸÜ⁄© ⁄ÜŸæ': 'Left Flank',
    'Ÿà€åŸÜ⁄Øÿ±': 'Winger',
    'ŸÅŸàÿ±Ÿàÿßÿ±ÿØ': 'Forward',
    'ÿØÿ±Ÿàÿßÿ≤Ÿá‚Äåÿ®ÿßŸÜ': 'Goalkeeper'
}
df['Position'] = df['Position'].map(position_translation)

# 4Ô∏è‚É£ Display First 5 Rows
print("First 5 rows of dataset:")
display(df.head())

# 5Ô∏è‚É£ Numeric Summary
print("\nNumeric summary of the dataset:")
display(df.describe())

# 6Ô∏è‚É£ Count of Players by Position
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Position', hue='Position', palette="viridis", dodge=False)
plt.title("Number of Players by Position")
plt.xlabel("Position")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend().remove()
plt.show()

# 7Ô∏è‚É£ Average Skills by Position
skill_columns = ['Speed','Agility','Endurance','Passing','Dribbling','Shooting']
avg_skills = df.groupby('Position')[skill_columns].mean()

plt.figure(figsize=(12,6))
avg_skills.plot(kind='bar', figsize=(12,6))
plt.title("Average Skills by Position")
plt.ylabel("Average Score")
plt.xlabel("Position")
plt.xticks(rotation=45)
plt.show()

# 8Ô∏è‚É£ Overall Skill Index
df['Skill_Index'] = df[skill_columns].mean(axis=1)

# Top 10 Players by Skill Index
top_players = df.sort_values('Skill_Index', ascending=False).head(10)
print("\nTop 10 Players by Skill Index:")
display(top_players[['Name','Position','Skill_Index'] + skill_columns])

# 9Ô∏è‚É£ Height vs Weight Scatter Plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Height_m', y='Weight_kg', hue='Position', s=100, palette='viridis')
plt.title("Height vs Weight of Players")
plt.xlabel("Height (m)")
plt.ylabel("Weight (kg)")
plt.legend(title='Position', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# üîü Height Distribution by Position
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x='Position', y='Height_m', palette='viridis')
plt.title("Height Distribution by Position")
plt.xlabel("Position")
plt.ylabel("Height (m)")
plt.xticks(rotation=45)
plt.show()

# 1Ô∏è‚É£1Ô∏è‚É£ Weight Distribution by Position
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x='Position', y='Weight_kg', palette='viridis')
plt.title("Weight Distribution by Position")
plt.xlabel("Position")
plt.ylabel("Weight (kg)")
plt.xticks(rotation=45)
plt.show()

# 1Ô∏è‚É£2Ô∏è‚É£ Professional Start Age Distribution
plt.figure(figsize=(10,5))
sns.histplot(df['Pro_Start_Age'], bins=8, kde=True, color='teal')
plt.title("Distribution of Professional Start Age")
plt.xlabel("Start Age")
plt.ylabel("Count")
plt.show()

# 1Ô∏è‚É£3Ô∏è‚É£ Skill Index Distribution
plt.figure(figsize=(10,5))
sns.histplot(df['Skill_Index'], bins=8, kde=True, color='orange')
plt.title("Distribution of Overall Skill Index")
plt.xlabel("Skill Index")
plt.ylabel("Count")
plt.show()

# üîÅ Save Dataset with Skill Index
os.makedirs("dataset", exist_ok=True)
df.to_csv("dataset/iran_futsal_players_with_skills.csv", index=False)
df.to_json("dataset/iran_futsal_players_with_skills.json", orient='records', force_ascii=False)

print("‚úÖ Dataset with Skill Index saved: CSV + JSON in 'dataset/' folder")
```
